# 1. Текстовое описание набора данных

В качестве набора данных мы будем использовать набор данных, содержащий рекомендации по диете RDA на основе индийской кухни - https://www.kaggle.com/datasets/tarunrm09/indian-cuisine-based-rda-diet-recommendation-data?resource=download

Эта информация может быть полезна для людей, следящих за своей фигурой, либо же старающихся избавиться от лишнего веса, поскольку на её основе можно выстроить сбалансированных рацион питания, а также правильно спланирваоть количество потребляемых в день бжу, каллорий и нужных микроэлементов.

Датасет включает в себя один файл: indian_rda_based_diet_recommendation_system.csv

Файл содержит в себе следующие колонки:
- Food_items - название продукта питания.
- Breakfast - 0, если продукт/блюдо не рекомендуется есть на завтрак, иначе - 1.
- Lunch - 0, если продукт/блюдо не рекомендуется есть на обед, иначе - 1.
- Dinner - 0, если продукт/блюдо не рекомендуется есть на ужин, иначе - 1.
- Veg nonveg - 0, если продукт/блюдо не является вегетерианским, иначе - 1.
- Calories - каллораж продукта/блюда.
- Fats - количество жиров в продукте/блюде.
- Proteins - количество белков в продукте/блюде.
- Iron -количество железа в продукте/блюде.
- Calcium - количество кальция в продукте/блюде.
- Sodium - количество натрия в продукте/блюде.
- Potassium - количество калия в продукте/блюде.
- Carbohydrates - количество углеводов в продукте/блюде.
- Fibre - количество клетчатки в продукте/блюде.
- Vitamins - количество витаминов в продукте/блюде.
- Sugars - количество сахаров в продукте/блюде.

# 4. Информация о корреляции признаков

```python
data_corr = data.drop(columns = 'Food_items')
```

```python
data_corr.head()
```

```python
data_corr.corr()
```

```python
fig, ax = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15,5))
sns.heatmap(data_corr.corr(method='pearson'), ax=ax[0])
sns.heatmap(data_corr.corr(method='kendall'), ax=ax[1])
sns.heatmap(data_corr.corr(method='spearman'), ax=ax[2])
fig.suptitle('Корреляционные матрицы, построенные различными методами')
ax[0].title.set_text('Pearson')
ax[1].title.set_text('Kendall')
ax[2].title.set_text('Spearman')
```

## Тепловая карта с указание размера

```python
fig, ax = plt.subplots(1, 1, sharex='col', sharey='row', figsize=(10,5))
fig.suptitle('Корреляционная матрица')
sns.heatmap(data_corr.corr(), cmap='YlGnBu', ax=ax)
```
